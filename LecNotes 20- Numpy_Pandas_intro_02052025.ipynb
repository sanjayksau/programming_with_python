- Fast and memory efficient.

##Numpy array creation:
- np.array
- np.arange
- np.linspace

---
- shape, dimension, size, data type
- reshape


```python
#np.array
#1-d array
arr = np.array([2, 3, 5, 7])
print("Array: ", arr)
print(f"Dimension: {arr.ndim}, Shape: {arr.shape}, Size: {arr.size}, Data Type: {arr.dtype}")
```

```python
#2-d array
arr2d = np.array([[2, 3, 5, 7], [3, 6, 8, 9]])
print("Array: \n", arr2d)
print(f"Dimension: {arr2d.ndim}, Shape: {arr2d.shape}, Size: {arr2d.size}, Data Type: {arr2d.dtype}")
```

```python
#np.arange
arr = np.arange(10)
print("Array: ", arr)
print(f"Dimension: {arr.ndim}, Shape: {arr.shape}, Size: {arr.size}, Data Type: {arr.dtype}")
```

```python
arr = np.arange(2, 16, 3)
print("Array: ", arr)
print(f"Dimension: {arr.ndim}, Shape: {arr.shape}, Size: {arr.size}, Data Type: {arr.dtype}")
```

```python
#2-d using arange and reshape
arr = np.arange(10).reshape(2, 5)
print("Array: \n", arr)
print(f"Dimension: {arr.ndim}, Shape: {arr.shape}, Size: {arr.size}, Data Type: {arr.dtype}")
```

##Indexing and Slicing

```python
arr = np.arange(12).reshape(3, 4)
print(arr)
```

```python
print(arr[1][2], arr[1, 2], arr[2, 3])
```

```python
print(arr[1:3, 1:3])
```

```python
#First 2 columns
print('last 2 columns: \n', arr[:, :2])
#last 2 rows
print('last 2 rows: \n', arr[1:, :])
#first and 3rd row
print('1st and 2nd row: \n', arr[[0, 2], :])
#2nd and 4th column
print('2nd and 4th column: \n', arr[:, [1, 3]])

#other combinations if time permits
#Exercise with shape and dimension:
```

## Maths operations

```python
arr = np.arange(6).reshape(2, 3)
print('arr: \n', arr)
```

```python
#Broadcasting
print('arr + 2: \n', arr + 2)
print('arr - 2: \n', arr - 2)
print('arr * 2: \n', arr * 2)
print('arr / 2: \n', arr / 2)
print('arr ** 0.5: \n', arr ** 0.5)
```

```python
#Transpose, element-wise multiplication and matrix multiplication
print('arr.T: \n', arr.T)
print('arr * arr: \n', arr * arr) #element-wise
print('arr @ arr.T: \n', arr @ arr.T) #Lazy example for regular matrix multiplication
```

```python
#row-wise sum, column-wise sum, sum
print('Array: \n', arr)
print('arr.sum(axis=0): \n', arr.sum(axis=0))
print('arr.sum(axis=1): \n', arr.sum(axis=1))
print('arr.sum(): \n', arr.sum())
```

```python
#mean, std, min, max, median
print('arr.mean(): \n', arr.mean())
print('arr.std(): \n', arr.std())
print('arr.min(): \n', arr.min())
print('arr.max(): \n', arr.max())
```

##Creating Series

```python
import pandas as pd
#creating series from lists
#Series: 1-dimensional labeled array
ser = pd.Series([2, 3, 5, 7])
print(ser)
```

```python
#index and column
print("Index: \n", ser.index)
print("Values: \n", ser.values)
```

##Creating DataFrame

```python
#DataFrame: 2-dimensional labeled data structure (like a table)
#Create a DataFrame
df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]})
print(df)
```

```python
#index and columns
print("Index: \n", df.index)
print("Columns: \n", df.columns)
```

```python
#indexing in DataFrame
print(df['Name']) #column indexing
```

```python
#row-indexing
print(df.loc[1])
```

```python
#loc and iloc difference with example
#custom label, default index
#loc for custom label, iloc for default index
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['a', 'b', 'c'])
print(df)
```

##loc and iloc

```python
#loc example
print(df.loc['b'])
#iloc example
print(df.iloc[1])
```

##Selecting Columns and Rows

```python
#selecting columns from df
print(df[['A', 'B']])
```

```python
#selecting rows from df
print(df.loc['a'])
```

```python
#indexing one entry in DataFrame
print('DataFrame: \n', df)
print('individual entres: ', df.loc['c', 'A'], df['A']['c']) #notice the difference
```

##Filtering and GroupBy

```python
#Create a DataFrame to be used for filtering, and groupby
import pandas as pd

sales_data = {
    'Region': ['North', 'South', 'North', 'South', 'North', 'South'],
    'Product': ['Apple', 'Apple', 'Banana', 'Banana', 'Apple', 'Banana'],
    'Sales': [120, 100, 90, 80, 150, 130],
    'Qty': [10, 8, 9, 7, 12, 10]
}

sales_df = pd.DataFrame(sales_data)
print(sales_df)
```

```python
#Filtering Data
# Filter rows where sales > 100
print(sales_df[sales_df['Sales'] > 100])
```

```python
#GroupBy Example
# Total sales by region
print(sales_df.groupby('Region')['Sales'].sum())

# Average quantity by product
print(sales_df.groupby('Product')['Qty'].mean())
```

##Head, Tail

```python
print(sales_df.head(3))       # First 3 rows
print(sales_df.tail(2))       # Last 2 rows
```

##describe, info
- describe: statistical summary of numerical columns
- info: shows DataFrame summary (number of non-null entries, DataTypes etc)

```python
print(sales_df.describe())
print("\n==================\n")
print(sales_df.info())
```

##Missing Values

```python
import pandas as pd
import numpy as np

data = {
    'Product': ['Apple', 'Banana', 'Apple', 'Banana', 'Apple'],
    'Sales': [120, 90, np.nan, 80, 150],
    'Qty': [10, 9, 12, np.nan, np.nan]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Detecting missing values
print(df.isnull())          # True where value is missing
print(df.isnull().sum())    # Count of missing values per column
```

```python
#Drop missing values
df_drop_rows = df.dropna()            # Drop rows with any missing value
df_drop_cols = df.dropna(axis=1)      # Drop columns with any missing value
print(df_drop_rows)
```

```python
#Fill Missing Value
df_fill_zero = df.fillna(0)           # Fill all missing values with 0

df_fill_mean = df.fillna({
    'Sales': df['Sales'].mean(),
    'Qty': df['Qty'].mean()
})
```
